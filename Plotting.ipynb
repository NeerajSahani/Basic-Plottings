```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as mplstyle
mplstyle.use(['dark_background', 'ggplot'])
```

## Simple plot

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,7,8,6,4,6,8]

plt.plot(x,y)

plt.xlabel("x-axis")   #setting x label
plt.ylabel("y-axis")   #setting x label
plt.title('graph')
```

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,7,8,6,4,6,8]
y1=[7,7,7,8,9,4,6,8]

plt.plot(x, y, '--', c='g', marker='o', label="graph-1") #label will not add directly
plt.plot(x, y1, ':', c='b', label="graph-2") #label will not add directly

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('graph')
plt.legend(loc='best') #to show label
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

### Sub Plot

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
```

## Scatter Plot

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,7,8,6,4,6,8]
y1=[7,2,7,3,5,6,7,9]

plt.scatter(x, y,  marker="*", s=158, label="circle", alpha=0.9) 
plt.scatter(x, y1, marker="o", s=120, label="star", alpha=0.6) 

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title('Scatter Plot')
plt.legend(loc='best') #to show label
```

```python
#Another method
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
    }

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# a-> x-coordinate, b-> y-coordinate, c-> color, d-> size
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
```

## Histogram

```python
l=[1,2,3,4,5,6,7,8,9,1,11,23,12,13,14,15,16,17,18,23,34,23,25,2728,38,28]
f=[1,10,20,30]
plt.hist(l, f, color='g', rwidth=.9)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title('Histograms')
```

# Bar Plot

```python
x=['india','unan','iran','goa','up','patna']
y=[9,8,2,4,7,3]
xi=[i for i in range(len(x))]

plt.bar(xi,y,tick_label=x, color='green')
plt.xlabel('Country')
plt.ylabel('GDP')
```

# Pie Plot

```python
x=['usa','india','china','japan','uk','france']
y=[9,8,2,4,5,7]
plt.pie(y,labels=x,autopct="%0.2f", explode=(0,0.2,0,0,0,0.1),shadow=True)
plt.xlabel('X -axis')
plt.ylabel('Y -axis')
plt.show()
```

# Stack Plot

```python
x  =[1, 2, 3, 4, 5]
y  =[4, 5, 8, 6, 2]
y1 =[9, 3, 2, 5, 4]
y2 =[4, 5, 7, 9, 6]

plt.stackplot(x, y, y1, y2, labels=['first', 'second', 'third'])
plt.xlabel('COUNTRY')
plt.ylabel('test')
plt.title('Stack Plot')
plt.legend()
```
